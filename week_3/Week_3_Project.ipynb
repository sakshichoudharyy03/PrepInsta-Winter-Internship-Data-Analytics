**Question 2. What are the data types of each indicator, and do they align with their expected types (e.g., numerical, categorical)?**

```python
# Display the data types of each column
data_types = data.dtypes
print("Data Types of Each Column:")

for column, dtype in data_types.items():
    print(f"{column}: {dtype}")

    if data[column].dtype == 'object':
        print(f"Expected Type: Categorical")
    elif data[column].dtype == 'float64':
        print(f"Expected Type: Numerical")


    print()
```

**Question 3. Identify potential outliers in numerical indicators (e.g., Age, Salary, Rating). Should outliers be removed or adjusted?**

```python
z_score_threshold = 3

z_scores = np.abs((data[['Age', 'Salary', 'Rating']] - data[['Age', 'Salary', 'Rating']].mean()) / data[['Age', 'Salary', 'Rating']].std())

data_no_outliers = data[(z_scores <= z_score_threshold).all(axis=1)]

print("Dataset after removing outliers:")
print(data_no_outliers)
```

**Question 4: Examine the format of the Salary column. Does it require any formatting or standardization for consistent analysis?**

```python
# Extracting the numeric values from the Salary column
data['Salary_min'] = data['Salary'].apply(lambda x: int(x.split('-')[0].strip('$').strip('k')))
data['Salary_max'] = data['Salary'].apply(lambda x: int(x.split('-')[1].strip('$').strip('k')))

# Calculating the average (mean) of the salary range
data['Average_Salary'] = (data['Salary_min'] + data['Salary_max']) / 2

# Displaying the updated DataFrame with numerical representation of salary ranges
print(data)
```

**Question 6. Explore the Established column. Are there any inconsistencies or anomalies that need to be addressed?**

```python
# Identify unique values and their counts
print(data['Established'].value_counts())

established_dtype = data['Established'].dtype
print(established_dtype)
```

```python
# Replace '-1' with NaN in the 'Established' column and convert it to numeric
data['Established'] = pd.to_numeric(data['Established'], errors='coerce')

# Calculate median excluding NaN values
median_value = np.nanmedian(data['Established'])

# Replace NaN (originally -1 values) with the calculated median
data['Established'] = data['Established'].fillna(median_value)

# Convert float values to integers by rounding
data['Established'] = data['Established'].astype(int)

# Display the updated 'Established' column without decimal points
print(data['Established'])

# Convert back to object type if needed
data['Established'] = data['Established'].astype(object)


# Display the updated 'Established' column
print(data['Established'])
```

**Question 7: Analyze the Easy Apply column. Does it contain boolean values or need transformation for better analysis?**

```python
# Replace '-1' with False in the 'Easy Apply' column
data['Easy Apply'].replace('-1', False, inplace=True)

# Convert 'TRUE' values to actual boolean True
data['Easy Apply'].replace('TRUE', True, inplace=True)

# Display the updated 'Easy Apply' column
print(data['Easy Apply'])
```

**Question 8: Investigate the range of values in the Rating column. Does it fall within expected rating scales, and how should outliers be treated?**

```python
data['Rating'].value_counts()
```

```python
# Replace '-1' with NaN in the 'Rating' column and convert it to numeric
data['Rating'] = pd.to_numeric(data['Rating'], errors='coerce')

# Calculate the median excluding NaN values
median_rating = np.nanmedian(data['Rating'])

# Replace ratings below 0 or above 5 with the median
data.loc[(data['Rating'] < 0) | (data['Rating'] > 5), 'Rating'] = median_rating

# Display the updated 'Rating' column
print(data['Rating'])
```
